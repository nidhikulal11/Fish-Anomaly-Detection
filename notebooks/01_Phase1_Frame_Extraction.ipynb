```python
import os, cv2
import pandas as pd
from pathlib import Path
from tqdm import tqdm

PROJECT_ROOT="/content/drive/MyDrive/FishProject"
INPUT_ROOT="/content/drive/MyDrive/Fish behaviour videos/White Oranda Grade AA Goldfish Carassius Auratus"
OUTPUT_ROOT=f"{PROJECT_ROOT}/extracted_frames"
METADATA_ROOT=f"{PROJECT_ROOT}/metadata"
REPORT_ROOT=f"{PROJECT_ROOT}/reports"
TARGET_FPS=5
VIDEO_EXTENSIONS=(".mp4",".avi",".mov",".mkv",".m4v")

def mount_drive():
    from google.colab import drive
    drive.mount('/content/drive')

def create_dirs():
    for p in [OUTPUT_ROOT,METADATA_ROOT,REPORT_ROOT]:
        os.makedirs(p,exist_ok=True)

def scan():
    vids=[]
    for e in VIDEO_EXTENSIONS:
        vids.extend(Path(INPUT_ROOT).rglob(f"*{e}"))
    return sorted(vids)

def extract(video):
    cat=video.parent.name.lower()
    out=Path(OUTPUT_ROOT)/cat/video.stem
    meta=Path(METADATA_ROOT)/cat
    out.mkdir(parents=True,exist_ok=True)
    meta.mkdir(parents=True,exist_ok=True)
    cap=cv2.VideoCapture(str(video))
    if not cap.isOpened():
        return {"video":video.stem,"category":cat,"status":"FAILED","frames":0}
    fps=cap.get(cv2.CAP_PROP_FPS)
    step=max(1,round(fps/TARGET_FPS))
    rows=[];idx=0;count=0
    while True:
        ok,frame=cap.read()
        if not ok: break
        if idx%step==0 and frame.mean()>5:
            name=f"frame_{idx:06d}.jpg"
            cv2.imwrite(str(out/name),frame,[cv2.IMWRITE_JPEG_QUALITY,95])
            rows.append({"video":video.stem,"category":cat,"original_frame":idx,
                         "timestamp_sec":round(idx/fps,3),"filename":name})
            count+=1
        idx+=1
    cap.release()
    pd.DataFrame(rows).to_csv(meta/f"{video.stem}_metadata.csv",index=False)
    return {"video":video.stem,"category":cat,"status":"SUCCESS","frames":count}

def main():
    mount_drive()
    create_dirs()
    videos=scan()
    print(f"Found {len(videos)} videos")
    summary=[]
    for v in tqdm(videos):
        summary.append(extract(v))
    pd.DataFrame(summary).to_csv(f"{REPORT_ROOT}/extraction_summary.csv",index=False)
    print("Done.")

if __name__=="__main__":
    main()
```

```python
import os
import cv2
import pandas as pd
from pathlib import Path
from tqdm import tqdm

# ==========================================================
# CONFIGURATION
# ==========================================================

PROJECT_ROOT = "/content/drive/MyDrive/FishProject"

INPUT_ROOT = "/content/drive/MyDrive/Fish behaviour videos/White Oranda Grade AA Goldfish Carassius Auratus"

OUTPUT_ROOT = f"{PROJECT_ROOT}/extracted_frames"
METADATA_ROOT = f"{PROJECT_ROOT}/metadata"
REPORT_ROOT = f"{PROJECT_ROOT}/reports"

TARGET_FPS = 5

VALID_EXTENSIONS = {
    ".mp4",
    ".avi",
    ".mov",
    ".mkv",
    ".m4v"
}


# ==========================================================
# GOOGLE DRIVE
# ==========================================================

def mount_drive():
    from google.colab import drive
    drive.mount('/content/drive', force_remount=False)


# ==========================================================
# CREATE OUTPUT DIRECTORIES
# ==========================================================

def create_directories():

    os.makedirs(OUTPUT_ROOT, exist_ok=True)
    os.makedirs(METADATA_ROOT, exist_ok=True)
    os.makedirs(REPORT_ROOT, exist_ok=True)

    print("Output folders created.")


# ==========================================================
# SCAN VIDEOS
# ==========================================================

def scan_videos():

    videos = []

    root = Path(INPUT_ROOT)

    for file in root.rglob("*"):

        if not file.is_file():
            continue

        if file.suffix.lower() not in VALID_EXTENSIONS:
            continue

        parent = file.parent.name.lower()

        # Only use videos inside healthy/unhealthy folders
        if parent not in ["healthy", "unhealthy"]:
            print(f"Skipping : {file}")
            continue

        videos.append(file)

    return sorted(videos)


# ==========================================================
# FRAME EXTRACTION
# ==========================================================

def extract_frames(video_path):

    category = video_path.parent.name.lower()

    video_name = video_path.stem

    save_folder = Path(OUTPUT_ROOT) / category / video_name
    save_folder.mkdir(parents=True, exist_ok=True)

    metadata_folder = Path(METADATA_ROOT) / category
    metadata_folder.mkdir(parents=True, exist_ok=True)

    cap = cv2.VideoCapture(str(video_path))

    if not cap.isOpened():

        return {
            "Video": video_name,
            "Category": category,
            "Status": "FAILED",
            "Frames": 0
        }

    fps = cap.get(cv2.CAP_PROP_FPS)

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if fps <= 0:
        fps = 30

    interval = max(1, round(fps / TARGET_FPS))

    metadata = []

    frame_number = 0

    saved_frames = 0

    while True:

        ret, frame = cap.read()

        if not ret:
            break

        if frame_number % interval == 0:

            # Skip black frames
            if frame.mean() < 5:
                frame_number += 1
                continue

            filename = f"frame_{frame_number:06d}.jpg"

            save_path = save_folder / filename

            cv2.imwrite(
                str(save_path),
                frame,
                [cv2.IMWRITE_JPEG_QUALITY, 95]
            )

            metadata.append({

                "video": video_name,

                "category": category,

                "original_frame": frame_number,

                "timestamp_sec": round(frame_number / fps, 3),

                "filename": filename

            })

            saved_frames += 1

        frame_number += 1

    cap.release()

    metadata_df = pd.DataFrame(metadata)

    metadata_df.to_csv(

        metadata_folder / f"{video_name}_metadata.csv",

        index=False

    )

    return {

        "Video": video_name,

        "Category": category,

        "Status": "SUCCESS",

        "Original FPS": round(fps,2),

        "Total Frames": total_frames,

        "Extracted Frames": saved_frames

    }


# ==========================================================
# MAIN
# ==========================================================

def main():

    mount_drive()

    create_directories()

    videos = scan_videos()

    print("=" * 60)
    print(f"Videos Found : {len(videos)}")
    print("=" * 60)

    if len(videos) == 0:

        print("No videos found.")
        return

    summary = []

    for video in tqdm(videos):

        info = extract_frames(video)

        summary.append(info)

    summary_df = pd.DataFrame(summary)

    summary_df.to_csv(

        f"{REPORT_ROOT}/extraction_summary.csv",

        index=False

    )

    print("\nExtraction Completed Successfully\n")

    print(summary_df)

    print("\nReports saved to:")

    print(REPORT_ROOT)


if __name__ == "__main__":
    main()
```
